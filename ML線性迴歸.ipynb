```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
# ys = 3xs + 1
xs = np.array([1.01, 0.49, 0.31, 1.51, 0.37, 0.73, 1.53, 0.56, 0.41, 0.74, 0.63,0.6,2.06,1.1,1.32,2.02,0.34], dtype=float)
ys = np.array([7.366, 0.985, 0.544, 9.140, 0.493, 3.011, 11.413, 1.814, 0.876, 2.690, 1.991, 4.172, 11.764, 4.682, 6.171, 15.996, 0.695], dtype=float)
# ys = 5xs + 2
# xs = np.array([6.0, 7.0, 8.0, 9.0, 10.0, 11.0], dtype=float)
# ys = np.array([32.0, 37.0, 42.0, 47.0, 52.0, 57.0], dtype=float)

model.fit(xs ,ys , epochs=1000)
model.save('hello_world_model.h5')
# creates a HDF5 file 'hello_world_model.h5'
print(model.predict([1.3]))
```
